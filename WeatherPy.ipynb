```python
#Initiate counter for city number below
number = 1

#Lists for holding required items
city_name = []
lat = []
lng = []
temp = []
humid = []
clouds = []
wind = []
country = []
date = []
max_temp = []

#Loop through each city, extract required items, and print result

print('Beginning Data Retrieval')
print('------------------------------')

for city in cities:
    try:
        city_data = (requests.get(url + city)).json()
        city_name.append(city_data['name'])
        lat.append(city_data['coord']['lat'])
        lng.append(city_data['coord']['lon'])
        temp.append(city_data['main']['temp'])
        humid.append(city_data['main']['humidity'])
        clouds.append(city_data['clouds']['all'])
        wind.append(city_data['wind']['speed'])
        country.append(city_data['sys']['country'])
        date.append(city_data['dt'])
        max_temp.append(city_data['main']['temp_max'])
        print(f'Processing Record {number} of {len(cities)} complete. | {city}')
        number += 1
    
    except KeyError:
        print(f'City weather not found. Skipping...')
        number += 1
        
print('------------------------------')
print('Data Retrieval Complete')
print('------------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Create a dataframe with required items
city_data_df = pd.DataFrame({'City': city_name,
                            'Lat': lat,
                            'Lng': lng,
                            'Max Temp': max_temp,
                            'Humidity': humid,
                            'Cloudiness': clouds,
                            'Wind Speed': wind,
                            'Country': country,
                            'Date': date})

#Save dataframe as csv
city_data_df.to_csv(output_data_file)

city_data_df.head()
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Cut city_data by northern hemisphere

nth_city_data_df = city_data_df.loc[city_data_df['Lat'] > 0]
# nth_city_data_df.head()

# Linear regression model for latitude and maximum temperature

nth_lat = nth_city_data_df.iloc[:,1]
nth_max_temp = nth_city_data_df.iloc[:,3]
nth_temp_r_value = st.pearsonr(nth_lat, nth_max_temp)
print(f"The r-value is = {nth_temp_r_value[0]}")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(nth_lat, nth_max_temp)

nth_temp_line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

nth_temp_regress_values = (slope * nth_lat) + intercept

# Generate a scatter plot of maximum temperature vs latitude (Northern Hemisphere)

plt.scatter(nth_city_data_df.iloc[:,1], nth_city_data_df.iloc[:,3], edgecolors='black')

plt.plot(nth_lat, nth_temp_regress_values,"r-")

plt.annotate(nth_temp_line_eq,(5,35),fontsize=12,color="red")
      
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('City Latitude vs. Max Temperature (Northern Hemisphere)')
plt.grid();
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Cut data by southern hemisphere

sth_city_data_df = city_data_df.loc[city_data_df['Lat'] < 0]
# sth_city_data_df.head()

# Linear regression model for latitude and maximum temperature

sth_lat = sth_city_data_df.iloc[:,1]
sth_max_temp = sth_city_data_df.iloc[:,3]
sth_temp_r_value = st.pearsonr(sth_lat, sth_max_temp)
print(f"The r-value is = {sth_temp_r_value[0]}")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(sth_lat, sth_max_temp)

sth_temp_line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

sth_temp_regress_values = (slope * sth_lat) + intercept

# Generate a scatter plot of maximum temperature vs latitude (Northern Hemisphere)

plt.scatter(sth_city_data_df.iloc[:,1], sth_city_data_df.iloc[:,3], edgecolors='black')

plt.plot(sth_lat, sth_temp_regress_values,"r-")

plt.annotate(sth_temp_line_eq,(-54,85),fontsize=12,color="red")

plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('City Latitude vs. Max Temperature (Southern Hemisphere)')
plt.grid();
```
